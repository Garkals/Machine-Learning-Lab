```python
a= int(input("Enter the number a:"))
b=int(input("Enter the number b:"))
ch=input("Enter the choice(+,-,*,/):")
if ch=="+":
     print(a+b)
elif ch=="-":
     print(a-b)
elif ch=="*":
     print(a*b)
elif ch=="/":
     if b > 0:
         print(a/b)
     else:
          print("enter the value of b great than 0")  
else:
    print("Enter the right choice")                



      
    
```
